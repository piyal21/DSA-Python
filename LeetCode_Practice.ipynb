```python
# fibonacci using recursion

def fiboncacci(num):
    if num == 0 or num ==1 :
        return num 
    
    return fiboncacci(num-1)+fiboncacci(num-2)


print('Enter a number to find the nth fibonacci number : ')
k= int(input())
fiboncacci(k)
```

```python
# two sum 

def sol(nums,target):
    
    new_dict={}
    
    for index,values in enumerate(nums):
        complement = target-values
        if complement in new_dict:
            return [new_dict[complement],index]
        new_dict[values]=index
        
k= [ 3,5,6,9,8]
sol(k,14)
```

```python
# -> Bubble sort / Adjacent sort[swap]
# ->Time complexity : O((n*(n+1))/2) or |  O(n^2) -> Avg/ Worst case
# -> Best case : O(n^2)


nums = [ 5,8,1,6,9,2,4,10]
print(len(nums))

def bubble_sort(nums):
    
    n =len(nums)
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if nums[j] > nums[j+1]:
                nums[j],nums[j+1]= nums[j+1],nums[j]
    print(nums)
    
bubble_sort(nums)
```

```python
# Optimized bubble sort 
# best case -> O(n)

nums = [ 5,8,1,6,9,2,4,10]
print(len(nums))

def bubble_sort(nums):
    
    n =len(nums)
    for i in range(n-2,-1,-1):
        swap_done = False
        for j in range(0,i+1):
            if nums[j] > nums[j+1]:
                nums[j],nums[j+1]= nums[j+1],nums[j]
            swap_done=True
            
        if swap_done ==False:
            break
    print(nums)
    
bubble_sort(nums)
```

```python
# insertion sort

nums = [ 3,5,6,4,8,9,10,7,1]
 


def insertion_sort(nums):
    n = len(nums)
    
    for i in range(1, n):
        key = nums[i]
        j = i -1 
        
        while j >=0 and nums[j] > key:
            nums [ j +1 ] = nums[j]
            j-=1
            
        nums[j+1]=key
        
        
insertion_sort(nums)
print(nums)
```

```python
# quick sorting algo


nums = [ 4,1,7,6,3,2,8]
low= 0
high = len(nums) - 1

def partition(nums,low,high):
  pivot = nums[low]
  i = low 
  j = high

  while i < j :
    while nums[i] <= pivot and i <= high-1:
      i+=1
    while nums[j] > pivot and j >= low +1:
      j-=1
    
    if i < j :
      nums[i],nums[j] = nums[j],nums[i]

  nums[low],nums[j]=nums[j],nums[low]

  return j


def quick_sort(nums,low,high):
  if low < high:
    p_index = partition(nums,low,high)
    quick_sort(nums,0,p_index-1)
    quick_sort(nums,p_index+1,high)

quick_sort(nums,low,high)
print(nums)
```

```python
# leetcode 268: finding missing number in an array 

nums = [9,6,4,2,3,5,7,0,1]

def finding_missing_values(nums):
    n = len(nums)
    freq = {}
    
    for i in range(0,n+1):
        freq[i]=0
        
    for i in nums:
        freq[i]=1
        
    for k,v in freq.items():
        if v ==0:
            return k
        
missing_value = finding_missing_values(nums)
print(missing_value)
```

```python
# leetcode 485 


nums =[1,1,0,1,1,1]



def max_consecutive_ones(nums):
    n = len(nums)
    
    count =0
    max_count = 0 
    
    for i in range(0,n):
        if nums[i] == 1:
            count+=1
        else:
            max_count = max(max_count,count)
            count= 0
            
    return max(max_count,count)

m_count = max_consecutive_ones(nums)
print(m_count)
```
